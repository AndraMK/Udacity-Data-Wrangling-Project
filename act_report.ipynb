## Report: act_report


This project gathered information from the Twitter webpage ‘We Rate Dogs’ which allows users to post images of their dogs and rate them. Although ratings should fall on a 10/10 rating scale, ratings falling outside of that scale are encouraged, primarily in the numerator. Information for this analysis included tweet id’s, tweet timestamps, webpage url’s, image url’s, dog ratings, an algorithm to determine dog breed, dog names, and information regarding likes, retweets, quotes, and responses regarding tweet posts on the We Rate Dogs twitter webpage. Analysis of this information yielded several insights. 


### Insight #1:

Significantly more people rated dogs at 10 or above. 
The proportion of dog ratings that fell below 10 was 1.51
The proportion of dog ratings equal to or above 10 was 10.87.

### Insight #2:

The top 10 most popular dog names reported in the 'We Rate Dogs' dataset are: Charlie, Cooper, Lucy, Oliver, Penny, Tucker, Winston, Sadie, Daisy, and Toby.



### Insight #3:

4. chihuahua
5. samoyed
6. french bulldog

### Insight #4:

The Average number of likes per tweet: 7580

The Average number of retweets per tweet: 2212



There appears to be a positive correlation between likes per tweet and retweets per tweet. As this graph suggests, tweets that receive more likes are more likely to be retweeted.
